```python
#importing libraries

import matplotlib.pyplot as plt
import pandas as pd
```

```python
#reading csv files

ayp = pd.read_csv("RICE APY - RICE APY.csv")
brkn_rice = pd.read_csv("BROKEN RICE MARKET PRICE_qt - BROKEN RICE MARKET PRICE.csv")
rice = pd.read_csv("RICE MARKET PRICE_qt - RICE MARKET PRICE.csv")
e_qty = pd.read_csv("EXPORT QTY - EXPORT QTY.csv")
```

YEAR VS AREA

```python
#plotting graph
plt.plot(ayp['Year'], ayp['Area (ha)'], marker='o')
plt.xlabel('YEAR')
plt.ylabel('AREA (ha)')
plt.title('YEARWISE CULTIVATION AREA')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

YEAR VS PRODUCTION

```python
#plotting graph
plt.plot(ayp['Year'], ayp['Production (10000 qt)'], marker='o')
plt.xlabel('YEAR')
plt.ylabel('PRODUCTION (10000 qt)')
plt.title('YEARWISE PRODUCTION')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

RICE AND BROKEN RICE VS YEAR

```python
#plotting graphs
plt.plot(rice['year'], rice['value'], label='RICE', marker='o')
plt.plot(brkn_rice['year'], brkn_rice['value'], label='BROKEN RICE', marker='o')

#giving labels to both axis
plt.xlabel('YEAR')
plt.ylabel('RICE MARKET PRICE (Rs./qt)')
plt.title('DOMESTIC MARKET RICE RETAIL PRICE')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

PRODUCTION QTY VS EXPORT QTY

```python
x = ayp['Year']
y1 = ayp['Production (10000 qt)']
y2 = e_qty['Total (Q)']

fig, ax = plt.subplots()

ax.plot(x, y1, color='blue')
ax2 = ax.twinx()
ax2.plot(x, y2, color='red')

ax.set_xlabel('YEARS')
ax.set_ylabel('PRODUCTION QUANTITY', color='blue')
ax2.set_ylabel('EXPORT QUANTITY', color='red')

# Legends for both plots
ax.legend(['PROD'], loc='upper left')
ax2.legend(['EXP'], loc='upper right')

plt.show()
```

DOMESTIC PRICE VS EXPORT PRICE

```python
x1 = rice['year']
y1_rice = rice['value']
y2_rice = e_qty['Total (P)']

fig, ax1 = plt.subplots()

ax1.plot(x1, y1_rice, color='blue', label='DOM')
ax2 = ax1.twinx()
ax2.plot(x1, y2_rice, color='red', label='EXP')

ax1.set_xlabel('YEARS')
ax1.set_ylabel('RICE PRICE (Rs./qt)', color='blue')
ax2.set_ylabel('EXPORT PRICE (Rs. Lakh/qt)', color='red')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()

# Plot for Broken Rice
x2 = brkn_rice['year']
y1_brkn_rice = brkn_rice['value']
y2_brkn_rice = e_qty['Total (P)']

fig, ax1 = plt.subplots()

ax1.plot(x2, y1_brkn_rice, color='blue', label='DOM')
ax2 = ax1.twinx()
ax2.plot(x2, y2_brkn_rice, color='red', label='EXP')

ax1.set_xlabel('YEARS')
ax1.set_ylabel('BROKEN RICE PRICE (Rs./qt)', color='blue')
ax2.set_ylabel('EXPORT PRICE (Rs. Lakh/qt)', color='red')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.show()
```

EXPORT VARIETY COMPARISION

```python
plt.plot(e_qty['Year'], e_qty['Non-Basmati Rice (Q)'], label='N0N-BASMATI', marker='o')
plt.plot(e_qty['Year'], e_qty['(Q)Other Rice Semi Milled Or Wholly Milled Rice, Whether Or Not Polished Or Glazed'], label='OTHER', marker='o')
plt.plot(e_qty['Year'], e_qty['Basmati Rice (Q)'], label='BASMATI', marker='o')

plt.xlabel('YEAR')
plt.ylabel('EXPORTED RICE QUANTITY (qt)')
plt.title('VARIETY EXPORT QUANTITY COMPARISION')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.plot(e_qty['Year'], e_qty['Non-Basmati Rice (P)'], label='N0N-BASMATI', marker='o')
plt.plot(e_qty['Year'], e_qty['(P)Other Rice Semi Milled Or Wholly Milled Rice, Whether Or Not Polished Or Glazed'], label='OTHER', marker='o')
plt.plot(e_qty['Year'], e_qty['Basmati Rice (P)'], label='BASMATI', marker='o')

plt.xlabel('YEAR')
plt.ylabel('EXPORTED RICE PRICE (Rs./qt)')
plt.title('VARIETY EXPORT PRICE COMPARISION')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.show()
```
